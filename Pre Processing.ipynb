```python
onehot = OneHotEncoder()
encoder_df = pd.DataFrame(onehot.fit_transform(df[['workpiece_slice_geometry']]).toarray())
df = df.join(encoder_df)
```

```python
df.rename(columns = {0:'150x150',1:'180x180'}, inplace = True)
```

```python
df = df.drop(['workpiece_slice_geometry'], axis=1)
```

```python
df['alloy_type'].unique()
```

```python
onehot = OneHotEncoder()
encoder_df = pd.DataFrame(onehot.fit_transform(df[['alloy_type']]).toarray())
df = df.join(encoder_df)
```

```python
df.rename(columns = {0:'close',1:'open'}, inplace = True)
```

```python
df = df.drop(['alloy_type'], axis=1)
```

```python
df['kind'].unique()
```

```python
onehot = OneHotEncoder()
encoder_df = pd.DataFrame(onehot.fit_transform(df[['kind']]).toarray())
df = df.join(encoder_df)
```

```python
df.rename(columns = {0:'B 789/BC',1:'SPH-B 795',2:'nan'}, inplace = True)
```

```python
df = df.drop(columns={'kind','nan'}, axis=1)
```

```python
df['time_temperature_measurement1'].unique()
```

```python
df['time_temperature_measurement2'].unique()
```

```python
df['sample_time_continuous_caster'].unique()
```

```python
df['sleeve'].unique()
```

```python
label_encoder = LabelEncoder()
df['time_temperature_measurement1']= label_encoder.fit_transform(df['time_temperature_measurement1']) 
df['time_temperature_measurement2']= label_encoder.fit_transform(df['time_temperature_measurement2']) 
df['sample_time_continuous_caster']= label_encoder.fit_transform(df['sample_time_continuous_caster']) 
df['sleeve']= label_encoder.fit_transform(df['sleeve'])
df['steel_type']= label_encoder.fit_transform(df['steel_type']) 
df['doc_requirement']= label_encoder.fit_transform(df['doc_requirement'])
```

```python
df.columns
```

```python
(((df.isnull().sum()/df.shape[0])*100)>0)&(((df.isnull().sum()/df.shape[0])*100)<10)
```

```python
df = df.dropna(subset=['grab1_num'])
df = df.dropna(subset=['steel_temperature_grab1, Celsius deg.'])
df = df.dropna(subset=['grab2_num'])
df = df.dropna(subset=['resistance, tonn'])
df = df.dropna(subset=['temperature_measurement1, Celsius deg.'])
df = df.dropna(subset=['temperature_measurement2, Celsius deg.'])
df = df.dropna(subset=['As, %'])
df = df.dropna(subset=['Mo, %'])
df = df.dropna(subset=['Nb, %'])
df = df.dropna(subset=['Ti, %'])
df = df.dropna(subset=['RUL'])
df = df.dropna(subset=['water_consumption_secondary_cooling_zone_num3, liter/minute'])
```

```python
((df.isnull().sum()/df.shape[0])*100)
```

```python
mean_value = df['quantity, tonn'].mean()
df['quantity, tonn'].fillna(value=mean_value, inplace=True)
```

```python
mean_value = df['Ce, %'].mean()
df['Ce, %'].fillna(value=mean_value, inplace=True)
```
